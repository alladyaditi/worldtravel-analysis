```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x="budget_level",y="seclusion",palette="flare")
plt.title("Culture Rating By Budget level")
plt.xlabel("Budget Level")
plt.ylabel("Seclusion")
plt.show
```

```python
numeric_columns = ["culture", "adventure", "nature", "beaches", "nightlife", "cuisine", "wellness", "urban", "seclusion"]
level = "Budget"   # choose: "budget", "mid-range", "luxury"
level_avg = df[df["budget_level"] == level][numeric_columns].mean()
level_avg
plt.figure(figsize=(9,9))
plt.pie(
    level_avg,
    labels=level_avg.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Spectral", len(level_avg))
)
plt.title(f"Travel Attribute Composition for {level.capitalize()} Destinations")
plt.show()

```

```python
 numeric_columns = ["culture", "adventure", "nature", "beaches", "nightlife", "cuisine", "wellness", "urban", "seclusion"]
level = "Luxury"   # choose: "budget", "mid-range", "luxury"
level_avg = df[df["budget_level"] == level][numeric_columns].mean()
level_avg
plt.figure(figsize=(9,9))
plt.pie(
    level_avg,
    labels=level_avg.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette("Spectral", len(level_avg))
)
plt.title(f"Travel Attribute Composition for {level.capitalize()} Destinations")
plt.show()
```

```python
for level in ["Budget","Luxury","Mid-range"]:
    level_avg = df[df["budget_level"] == level][numeric_columns].mean()
    plt.figure(figsize=(9,9))
    plt.pie(
        level_avg,
        labels=level_avg.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=sns.color_palette("Spectral", len(level_avg))
    )
    plt.title(f"Travel attribute Composition for {level.capitalize()} Destinations")
    plt.show()
```

```python
df["travel_score"]= df["culture"]+df["nightlife"]+df["cuisine"]+df["beaches"]+df["wellness"]+df["nature"]+df["urban"]+df["adventure"]+df["seclusion"]/9
 
```

# Conclusion

The analysis shows that **Thailand, Colombia, and the Philippines** are among the most budget-friendly travel destinations. Many cities within Thailand offer consistently high scores in **culture, cuisine, and nature**, making it one of the strongest low-cost regions for travelers seeking rich experiences without a high price tag.

In contrast, the most **luxury-oriented destinations** include **Canada, the United States, and Australia**. Cities across Oceania—especially **Hawaii and major cities in Australia**—stand out as premium travel locations. These luxury destinations score highly in **culture, cuisine, and nature**, providing a diverse range of high-quality travel experiences.

Overall, the results highlight clear differences between budget and luxury travel regions:

- **Budget destinations** offer exceptional cultural and natural value at lower costs.  
- **Luxury destinations** provide elevated experiences across similar key attributes, often with more diverse and premium travel offerings.
